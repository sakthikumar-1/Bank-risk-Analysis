```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Bank_analysis.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean(numeric_only= True))
```

```python
df.isnull().sum()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
    .str.replace('-','_')
)
```

```python
df['years_in_current_job'] = df['years_in_current_job'].fillna("Unknown")
df['home_ownership']= df['home_ownership'].fillna("Unknown")
df[['purpose','loan_status','term']] = df[['purpose','loan_status','term']].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df['years_in_current_job'] = df['years_in_current_job'].fillna(df['years_in_current_job'].mean())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df[df['current_loan_amount'] != 99999999.0]
```

```python
df = df.drop(columns=["loan_id", "customer_id"])
```

```python
def fix_years(x):
    if x == '10+ years': return 10
    if x == '< 1 year': return 0.5
    if x == 'Unknown': return None
    return float(x.split()[0])

df['years_in_current_job'] = df['years_in_current_job'].apply(fix_years)
```

```python
df.to_csv("Cleaned_Bank_risk.csv")
```
